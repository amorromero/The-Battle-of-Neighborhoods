## Bussines Problem

The objective of this project is to use Foursquare location data and regional clustering of venue information to determine what might be the ‘best’ neighbourhood in Toronto to open a Coffee Shop. This city is home to many coffee shops where coffee addicts find their favorite drink. Whether you’re on your way to work looking for a quick cup or a tourist fueling up while visiting the best attractions and parks or you’re simply looking for a spot to sit down and map out the best things to do during your local adventures, Toronto is one of the best options to drink a very good coffee.

But...

<b> will we find the most suitable location for an entrepreneur to open a new Coffee Shop in Toronto?</b>

## Data Overview

The data that will be required will be a combination of CSV files that have been prepared for the purposes of the analysis from multiple sources which will provide the list of neighbourhoods in Toronto (via Wikipedia), the Geographical location of the neighbourhoods (via Geocoder package) and Venue data pertaining to Coffee Shops (via Foursquare). The Venue data will help find which neighbourhood is best suitable to open an Coffee Shop.
